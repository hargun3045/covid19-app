```python
sns.set(rc = {'figure.figsize': (16,8)})
ax = sns.violinplot(x = 'urgency_of_admission', y = 'age', data = df)
```

```python
fig, ax = plt.subplots()
width = 0.4
fig.set_size_inches(16, 8)
(df[df.urgency_of_admission == 'Low'].age // 10 * 10).value_counts(normalize = True).sort_index().plot(label = 'Low Urgency', kind = 'bar',color = 'Blue', position = 0, ax = ax,width = width )
ax.legend(loc = 'upper left')
ax2 = ax.twinx()
(df[df.urgency_of_admission == 'High'].age // 10 * 10).value_counts(normalize = True).sort_index().plot(label = 'High Urgency', kind = 'bar',color = 'Red', position = 1, ax = ax2,width = width )
ax2.legend(loc = 'upper right')
```

```python
fig, ax = plt.subplots()
width = 0.4
fig.set_size_inches(16, 8)
(df[(df.sex == 'male') & (df.urgency_of_admission == 'Low')].age // 10 * 10).value_counts(normalize = True).sort_index().plot(label = 'Low Urgency', kind = 'bar',color = 'Blue', position = 0, ax = ax,width = width )
ax.legend(loc = 'upper left')
ax2 = ax.twinx()
(df[(df.sex = 'male') & (df.urgency_of_admission == 'High')].age // 10 * 10).value_counts(normalize = True).sort_index().plot(label = 'High Urgency', kind = 'bar',color = 'Red', position = 1, ax = ax2,width = width )
ax2.legend(loc = 'upper right')
```
